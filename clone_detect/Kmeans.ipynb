```python
# K-means clustering, silhouette score, and distortion

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from sklearn.metrics import silhouette_score

max_cluster = 15

pair_latent = squareform(pdist(latent))
score = []
labels = []
distortion = []

for m in np.arange(2, max_cluster + 1):

    kmeans = KMeans(n_clusters = m).fit(latent)
    labels.append(kmeans.labels_)
    score.append(silhouette_score(pair_latent, kmeans.labels_, metric = "precomputed"))
    distortion.append(kmeans.inertia_)
    
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(12, 5)
axs[0].plot(np.arange(2, max_cluster + 1), distortion)
axs[0].set_title("Distortion")
axs[0].set_xlabel("Number of Clusters")
axs[1].set_title("Silhouette Score")
axs[1].set_xlabel("Number of Clusters")
axs[1].plot(np.arange(2, max_cluster + 1), score)
plt.show()
```

```python
# Cluster labels and silhouette plot

import matplotlib.cm as cm 
from sklearn.metrics import silhouette_samples

cluster_no = 4
cluster = []
label = labels[cluster_no - 2]

for g in range(cluster_no):
    
    cluster.append(np.where(label == g)[0])
    
colors = cm.rainbow(np.linspace(0, 1, cluster_no))
silhouette_avg = silhouette_score(pair_latent, label, metric = "precomputed")
sample_silhouette_values = silhouette_samples(pair_latent, label, metric = "precomputed")

fig, ax1 = plt.subplots(1, 1)
fig.set_size_inches(6, 5)
ax1.set_xlim([-0.1, 1])
ax1.set_ylim([0, len(label) + (cluster_no + 1) * 10])
y_lower = 10

for i in range(cluster_no):

    ith_cluster_silhouette_values = sample_silhouette_values[label == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    
    y_upper = y_lower + size_cluster_i 

    ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values, facecolor = colors[i], edgecolor = colors[i], alpha = 0.7)
    ax1.text(- 0.05, y_lower + 0.5 * size_cluster_i, str(i + 1))
    
    y_lower = y_upper + 10
    
    ax1.set_title("silhouette plot", fontsize = 30, y = 1.05)
    ax1.set_xlabel("silhouette coefficient", fontsize = 15)
    ax1.set_ylabel("cluster", fontsize = 15)
    ax1.axvline(x = silhouette_avg, color = "red", linestyle = "--")
    ax1.set_yticks([])  
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

plt.show()
```
